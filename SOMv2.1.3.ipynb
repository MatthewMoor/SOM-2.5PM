```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import somoclu
%matplotlib inline 
```

```python
# Our data
dataset = pd.read_csv('BeijingPM20100101_20151231.csv')
X = dataset.iloc[:, 1:]
del X['cbwd']

X = X.dropna()
print(X)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
X = sc.fit_transform(X)
print(np.shape(X))
```

```python
n_rows, n_columns = 690,690
%time som = somoclu.Somoclu(n_columns, n_rows, data=X, maptype="toroid")
som.train(epochs=10, radius0=395, scale0=0.1, scaleN=0.01)
```

```python
som.view_umatrix()
som.view_component_planes()
```

```python
som.cluster()
som.view_umatrix(bestmatches=True)
```
