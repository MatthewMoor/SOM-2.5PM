```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import somoclu
%matplotlib inline 
```

```python
# Our data
data_bei = pd.read_csv('BeijingPM2.csv')
data_chen = pd.read_csv('ChengduPM20.csv')
data_guan = pd.read_csv('GuangzhouPM20.csv')
data_shan = pd.read_csv('ShanghaiPM20.csv')
data_shen = pd.read_csv('ShenyangPM20.csv')

datasets = [
    
    data_bei,
    data_chen,
    data_guan,
    data_shan,
    data_shen,
    
]

# describes our datasets

for i, u in enumerate(datasets):
    datasets[i] = u.dropna()
#     print('\ndatasets {}: {}'.format(i, u))
        
for i, data in enumerate(datasets):
    pollution_first = datasets[i].iloc[:, 6]
    pollution_second = datasets[i].iloc[:, 7]
    pollution_third = datasets[i].iloc[:, 8]
    PM = data.iloc[:, 6:9]
    print(pollution_first)
    
    common_polution = (pollution_first + pollution_second + pollution_third) / 3
#     datasets[i].drop(['cbwd'], axis=1)
    
    
#     for j in data:
#         if j == 6 or j == 7 or j == 8:
#             print(datasets[j])
            
    datasets[i]['average_PM'] = common_polution
#     print(datasets[i])

#     data['PM2'] = amount
#     print(data.head())
    
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
X = datasets[0].iloc[:, 1:]
del X['cbwd']
data = sc.fit_transform(X)
print(np.shape(X))
```

```python
n_rows, n_columns = 25,25
som = somoclu.Somoclu(n_columns, n_rows, data=X, maptype="planar")
som.train(epochs=1000,radius0=20, radiusN=1,radiuscooling="exponential",scale0=0.1, scaleN=0.035, scalecooling="exponential")
```

```python
som.view_umatrix()
som.view_component_planes()
```

```python
som.cluster()
som.view_umatrix(bestmatches=True)
```
