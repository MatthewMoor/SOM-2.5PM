```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import somoclu
%matplotlib inline 
```

```python
# Our data
dataset = pd.read_csv('dust_calculation.csv')
X = dataset.iloc[:, 1:]

from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
X = sc.fit_transform(X)
print(np.shape(X))
```

```python
n_rows, n_columns = 242, 242
som = somoclu.Somoclu(n_columns, n_rows, data=X, maptype="planar")
%time som.train(epochs=2000, radius0=121, radiusN=1, scale0=0.1, scaleN=0.01)
som.view_umatrix()

som.view_component_planes()
```

```python
som.cluster()
som.view_umatrix(bestmatches=True)
```
